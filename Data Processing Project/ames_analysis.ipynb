#### 2.2.1 Explanation of Missing Values

    - Fence is missing because there is no fence for the corresponding house

- Therefore, these are not true missing values but meaningful indicators of absence.

#### 2.2.3 Missingness Mechanism

##### Missingness Mechanism Classification

- LotFrontage → MAR (Missing at Random)
  - The missing values in LotFrontage depend on the Neighborhood.
  - Since LotFrontage varies across neighborhoods, the missingness is related to another observed variable (Neighborhood), making it MAR.

- PoolQC → MNAR (Missing Not at Random)
  - PoolQC is missing when a house does not have a pool.
  - The missingness is directly related to the absence of the feature itself, not random, making it MNAR.

- Electrical → MCAR (Missing Completely at Random)
  - The missing values in Electrical appear to be very few and do not show any clear pattern or dependency on other variables.
  - Therefore, it can be considered as missing completely at random.

## 3.2 Statistical Outlier Detection

### Z-Score Method

```python
sale_z = (df['SalePrice'] - df['SalePrice'].mean()) / df['SalePrice'].std()

grliv_z = (df['GrLivArea'] - df['GrLivArea'].mean()) / df['GrLivArea'].std()


outliers = df[(abs(sale_z) > 3) | (abs(grliv_z) > 3)]

outliers
```

```python
outliers.shape[0]
```

#### 3.2.2 IQR Method

```python
Q1 = df['LotArea'].quantile(0.25)
Q3 = df['LotArea'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['LotArea'] > upper_bound) | (df['LotArea'] < lower_bound)]
outliers
```

```python
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Number of outliers:", outliers.shape[0])
```

#### 3.2.4 Comparison of Z-score and IQR Methods

- The Z-score method detected 31 outliers in the SalePrice and GrLivArea columns, while the IQR method detected 69 outliers in the LotArea column.
- The IQR method identified more outliers compared to the Z-score method.

- The Z-score method assumes that the data follows a normal distribution.
- And it is sensitive to extreme values. It may fail to detect outliers in skewed data.

- In contrast, the IQR method does not assume normality.
- And is more appropriate to skewed distributions and extreme values.

- Since variables like LotArea are highly skewed in this dataset, the IQR method is more appropriate for detecting outliers.

- Therefore, the IQR method is recommended for this dataset as it provides a more reliable identification of outliers in skewed data.

### 3.3.3 Remove outliers vs Capping


- Removing outliers is appropriate,
     - When the outliers are likely due to data entry errors, measurement errors, or anomalies that do not represent real-world situations.
     - The number of outliers is very small and their removal does not significantly reduce the dataset size.

- Capping or transforming outliers is appropriate when,
  - The outliers are valid observations and contain meaningful information.
  - Removing them would result in loss of important data or reduce model generalization.

### 4.2.3 Label Encoding vs One-Hot Encoding

- Label encoding converts categorical values into numerical labels.
- It is appropriate for ordinal categorical variables where there is a natural order or ranking among the categories.

- One-hot encoding, creates separate binary columns for each category.
- It is suitable for nominal categorical variables where there is no inherent order among categories.

- Using label encoding on nominal data can introduce incorrect relationships, as the model may interpret the labels as having a meaningful order.
- One-hot encoding avoids this issue by treating each category independently.

- In summary, label encoding is used for ordered categories, while one-hot encoding is used for unordered categories.

### 4.4.4 Min-Max Scaling vs Standard Scaling

- Min-Max scaling transforms data into a fixed range, usually between 0 and 1.
- It is useful when the model requires bounded input values or when the distribution does not need to be normalized.

- Standard scaling transforms data to have a mean of 0 and a standard deviation of 1.
- It is useful when the model assumes normally distributed data or is sensitive to feature variance.

- Min-Max scaling is sensitive to outliers, as extreme values affect the scaling range.
- Standard scaling is more robust to outliers because it centers the data based on mean and variance.

#### When to use Min-Max Scaling:
- When using models like Neural Networks or K-Nearest Neighbors, where features need to be in a specific range.
- Example: Scaling pixel values (0–255) in image data to 0–1 for a neural network.

#### When to use Standard Scaling:
- When using models like Linear Regression, Logistic Regression, or Support Vector Machines, which assume normally distributed features.
- Example: Standardizing house price and area features before training a regression model to ensure equal contribution of features.

- In summary, Min-Max scaling is preferred when a bounded range is required, while Standard scaling is preferred when data needs to be normalized around the mean.

###  5.3.4 Multicollinearity

```python
high_corr = []

for i in corr_matrix.columns:
    for j in corr_matrix.columns:
        if i != j and abs(corr_matrix.loc[i, j]) > 0.8:
            pair = tuple(sorted([i, j]))
            if pair not in high_corr:
                high_corr.append(pair)

high_corr[:10]
```

```python
for pair in high_corr[:5]:
    print(pair, corr_matrix.loc[pair[0], pair[1]])
```

- Several pairs of features show high correlation (r > 0.8), indicating multicollinearity.
- For example:
  - GarageCars and GarageArea are also strongly correlated, since larger garages can accommodate more cars.

- Multicollinearity can be problematic in regression models because:
  - It makes it difficult to determine the individual effect of each feature on the target variable.
  - It can lead to unstable coefficient estimates, where small changes in data cause large changes in model parameters.
  - It may reduce model interpretability and increase variance.

- Therefore, highly correlated features should be handled carefully, for example by removing or combining them.

### 5.3.5 Spearman vs Pearson

```python
pearson_corr = df['OverallQual'].corr(df['SalePrice'], method='pearson')

spearman_corr = df['OverallQual'].corr(df['SalePrice'], method='spearman')

print("Pearson correlation:", pearson_corr)
print("Spearman correlation:", spearman_corr)
```


- The Pearson correlation measures the linear relationship between OverallQual and SalePrice.
- The Spearman correlation measures the monotonic relationship based on ranked values.

- Since OverallQual is an ordinal variable, Spearman correlation is often more appropriate because it considers the ranking of values.

- The two correlation values may differ because Pearson assumes a linear relationship and is sensitive to outliers, while Spearman is based on ranks and is more robust to non-linear relationships and outliers.
